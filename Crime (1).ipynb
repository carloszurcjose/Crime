```python
import pandas as pd

# Load your dataset
df = pd.read_csv('/content/drive/MyDrive/CodeQuantum/CrimeDataset.csv')

# === STEP 1: Define crime-to-severity mapping ===
crime_to_severity = {
    #  Severe Crimes
    'CRIMINAL HOMICIDE': 'Severe',
    'RAPE, FORCIBLE': 'Severe',
    'RAPE, ATTEMPTED': 'Severe',
    'CHILD ABUSE (PHYSICAL) - AGGRAVATED ASSAULT': 'Severe',
    'CHILD ABUSE (PHYSICAL) - SIMPLE ASSAULT': 'Severe',
    'SEXUAL PENETRATION W/FOREIGN OBJECT': 'Severe',
    'ORAL COPULATION': 'Severe',
    'SODOMY/SEXUAL CONTACT B/W PENIS OF ONE PERS TO ANUS OTH': 'Severe',
    'CHILD PORNOGRAPHY': 'Severe',
    'KIDNAPPING': 'Severe',
    'KIDNAPPING - GRAND ATTEMPT': 'Severe',
    'ASSAULT WITH DEADLY WEAPON, AGGRAVATED ASSAULT': 'Severe',
    'ASSAULT WITH DEADLY WEAPON ON POLICE OFFICER': 'Severe',

    #  High Severity
    'ROBBERY': 'High',
    'ATTEMPTED ROBBERY': 'High',
    'INTIMATE PARTNER - AGGRAVATED ASSAULT': 'High',
    'BURGLARY': 'High',
    'BURGLARY FROM VEHICLE': 'High',
    'BURGLARY, ATTEMPTED': 'High',
    'BURGLARY FROM VEHICLE, ATTEMPTED': 'High',
    'THEFT-GRAND ($950.01 & OVER)EXCPT,GUNS,FOWL,LIVESTK,PROD': 'High',
    'THEFT FROM MOTOR VEHICLE - GRAND ($950.01 AND OVER)': 'High',
    'WEAPONS POSSESSION/BOMBING': 'High',
    'DISCHARGE FIREARMS/SHOTS FIRED': 'High',
    'SHOTS FIRED AT INHABITED DWELLING': 'High',
    'BRANDISH WEAPON': 'High',

    #  Moderate Severity
    'BATTERY - SIMPLE ASSAULT': 'Moderate',
    'INTIMATE PARTNER - SIMPLE ASSAULT': 'Moderate',
    'THEFT PLAIN - PETTY ($950 & UNDER)': 'Moderate',
    'THEFT FROM MOTOR VEHICLE - PETTY ($950 & UNDER)': 'Moderate',
    'SHOPLIFTING-GRAND THEFT ($950.01 & OVER)': 'Moderate',
    'SHOPLIFTING - PETTY THEFT ($950 & UNDER)': 'Moderate',
    'VANDALISM - FELONY ($400 & OVER, ALL CHURCH VANDALISMS)': 'Moderate',
    'VANDALISM - MISDEAMEANOR ($399 OR UNDER)': 'Moderate',
    'TRESPASSING': 'Moderate',
    'BATTERY POLICE (SIMPLE)': 'Moderate',
    'VEHICLE - STOLEN': 'Moderate',
    'VEHICLE - ATTEMPT STOLEN': 'Moderate',

    #  Low Severity
    'DISTURBING THE PEACE': 'Low',
    'RESISTING ARREST': 'Low',
    'FAILURE TO YIELD': 'Low',
    'FALSE POLICE REPORT': 'Low',
    'PROWLER': 'Low',
    'UNAUTHORIZED COMPUTER ACCESS': 'Low',
    'CONTEMPT OF COURT': 'Low',
    'COUNTERFEIT': 'Low',
    'DOCUMENT FORGERY / STOLEN FELONY': 'Low',
    'DISHONEST EMPLOYEE - PETTY THEFT': 'Low',
    'CREDIT CARDS, FRAUD USE ($950 & UNDER': 'Low',
    'PICKPOCKET': 'Low',
    'PETTY THEFT - AUTO REPAIR': 'Low',
    'LOITERING': 'Low',
    'SHOPLIFTING - ATTEMPT': 'Low',
    'DISRUPT SCHOOL': 'Low'
}

# === STEP 2: Map severity levels to the crimes ===
df['Severity'] = df['Crm Cd Desc'].map(crime_to_severity).fillna('Uncategorized')

# === STEP 3: Group by AREA and Severity ===
grouped = df.groupby(['AREA NAME', 'Severity']).size().reset_index(name='Count')

# Sort for easier reading
grouped = grouped.sort_values(['AREA NAME', 'Count'], ascending=[True, False])

# === STEP 4: Print grouped output per area ===
for area in grouped['AREA NAME'].unique():
    print(f"\n=== {area.upper()} ===")
    area_group = grouped[grouped['AREA NAME'] == area]
    for _, row in area_group.iterrows():
        print(f" - {row['Severity']}: {row['Count']}")
```

```python
# Define severity-to-points mapping
severity_points = {
    'Severe': 15,
    'High': 10,
    'Moderate': 5,
    'Low': 1,
    'Uncategorized': 0  # Optional: can ignore or include as 0
}

# Add a column with weighted points
grouped['Points'] = grouped['Severity'].map(severity_points) * grouped['Count']

# Now compute total score per AREA NAME
severity_scores = grouped.groupby('AREA NAME')['Points'].sum().reset_index(name='Severity Score')

# Sort by score descending (most dangerous areas first)
severity_scores = severity_scores.sort_values(by='Severity Score', ascending=False)

# Print results
print(severity_scores)
```
